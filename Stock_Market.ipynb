```python
import pandas as pd
```

```python
apple = pd.read_csv('AAPL.csv')
google = pd.read_csv('GOOGL.csv')
microsoft = pd.read_csv('MSFT.csv')
tesla = pd.read_csv('TSLA.csv')
nvidia = pd.read_csv('NVDA.csv')
amazon = pd.read_csv('AMZN.csv')
```

```python
apple.head()
```

```python
amazon.head()
```

```python
dfs = [apple, google, microsoft, tesla, nvidia, amazon]
```

```python
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
apple.head(50)
```

```python
microsoft.head(200)
```

```python
for df in dfs:
    df['previous_day_closing_price'] = df.Close.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['change_in_price'] = df['Close'] - df['previous_day_closing_price']
```

```python
nvidia.head()
```

```python
for df in dfs:
    df['percent_change_in_price'] = df.Close.pct_change()
```

```python
tesla.head()
```

```python
for df in dfs:
    df['previous_day_volume'] = df.Volume.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['change_in_volume'] = df['Volume'] - df['previous_day_volume']
```

```python
for df in dfs:
    df['percent_change_in_volume'] = df.Volume.pct_change()
```

```python
apple.to_csv('Apple.csv')
google.to_csv('Google.csv')
nvidia.to_csv('NVIDIA.csv')
microsoft.to_csv('Microsoft.csv')
tesla.to_csv('Tesla.csv')
amazon.to_csv('Amazon.csv')
```
